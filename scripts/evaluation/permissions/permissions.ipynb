```python
def load_app_data(path_name, name):
    result = []
    for f_name in os.listdir(path_name):
        #print(f_name[0: len(f_name) -5])
        if f_name[0: len(f_name) -5] in mapping and mapping[f_name[0: len(f_name) -5]] != name:
            continue
        if name == "gp_2022" and (f_name[0: len(f_name) -5] in manual or f_name[0: len(f_name) -5] in applist):
            continue
        
        with open(path_name +"/" +f_name, "r") as f:
            
            result.append(json.load(f))
    
    return result
        
            
```

```python
def count__permission(name, dataset):
    count = 0
    for a in dataset:
        for p in a:
            if name.lower() in p.lower():
                count = count +1
                
    return count
```

```python
def get_all_matching_permission(dataset, names):
    result = set()
    for a in dataset:
        for p in a:
            for n in names:
                if n.lower() in p.lower():
                    result.add(p)
                    
    return result
```

```python
def has_permission(app, name):
    for p in app:
        if name.lower() in p.lower():
            return True
    return False

def count_apps_with_permission(name, dataset):
    count = 0
    for a in dataset:
        if has_permission(a, name):
            count = count +1
    return count
```

```python


def count_apps_with_both_permission(name1, name2, dataset):
    count = 0
    for a in dataset:
        if has_permission(a, name1) and has_permission(a, name2):
            count = count +1
    return count
```

```python


def count_apps_with_multiple_permission(names, dataset):
    count = 0
    for a in dataset:
        has_all  = True
        for name in names:
            if not has_permission(a, name) :
                has_all = False
        
        if has_all:
            count = count +1
    return count
```

```python
def count_permission(dataset):
    result = {}
    for a in dataset:
        test_set = set(a)
        if len(a) != len(test_set):
            print("different")
        for p in test_set:
            result[p] = result.get(p, 0) + 1
    return result
```

```python
def get_number_of_permission(dataset):
    result = []
    for a in dataset:
        result.append(len(a))
    
    return result
```

```python
def all_permissions():
    result = []
    for p in permission_file["permissions"]:
        #print(p)
        result.append(p)
    return result
```

```python
def permission_with_level(level):
    result = []
    for p in permission_file["permissions"]:
        protextion_level_splitted = permission_file["permissions"][p]["protectionLevel"].split("|")
        if level in protextion_level_splitted:
            result.append(p)
    return result
```

```python
all_p = all_permissions()
```

```python
p_dangerous = permission_with_level("dangerous")
```

```python
p_priv = permission_with_level("privileged")
```

```python
def has_one_matching(matching_permission, app):
    for p in app:
        if p in matching_permission:
            return True
    return False

def get_at_least_one(matching_permissions, dataset):
    count = 0
    for a in dataset:
        if has_one_matching(matching_permissions, a):
            count = count +1
    return count
```

```python
def has_one_not_matching_custom(a):
    for p in a:
        if not p.startswith("android.permission.") and not p.startswith("com.android."):
            return True
    return False

def has_one_not_matching( matching_permission, a):
    for p in a:
        if p not in matching_permission:
            return True
    return False

def get_at_least_one_non_android(dataset):
    count = 0
    for a in dataset:
        if  has_one_not_matching(all_p, a):
            count = count +1
    return count
```

```python


def get_at_least_one_non_android_custom(dataset):
    count = 0
    for a in dataset:
        if has_one_not_matching_custom(a):
            count = count +1
    return count
```

```python
def filter_permission_result_map(result_map, matching_permission):
    result = {}
    for k,v in result_map.items():
        if k in matching_permission:
            result[k] = v
    return result
```

```python
def filter_permission_result_map_non_standard_custom(result_map):
    result = {}
    for k,v in result_map.items():
        if not k.startswith("android.permission.") and not k.startswith("com.android."):
            result[k] = v
    return result
```

```python
def filter_permission_result_map_non_standard(result_map):
    result = {}
    for k,v in result_map.items():
        if k not in all_p:
            result[k] = v
    return result
```

```python
def sort_map(map_to_sort):
    vk_pairs = ((value, key) for (key,value) in map_to_sort.items())
    return sorted(vk_pairs, reverse=True)
```
